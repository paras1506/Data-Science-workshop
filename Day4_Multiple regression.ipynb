```python
import pandas as pd
```

```python
df = pd.read_csv('mtcars.csv')
```

```python
df
```

```python
 # seperate input and output
    
x = df[['disp', 'hp' ,'wt']]

y = df['mpg']
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(x,y)
```

```python
disp = 221
hp = 102
wt = 2.91

reg.predict([[disp,hp,wt]])
```

```python
reg.score(x,y)
```
