#### OUtlier Detection and removal

```python
from sklearn.ensemble import IsolationForest
iso = IsolationForest(contamination=0.05, random_state=0)
```

```python
clean = iso.fit_predict(x,y)
```

```python
x_clean = x[clean==1]
y_clean = y[clean==1]
```

```python
x_clean.shape, y_clean.shape
```

### Feature Selection

```python
from sklearn.feature_selection import SelectKBest, chi2
```

```python
skf = SelectKBest(k=5, score_func= chi2)
```

```python
skf.fit(x_clean,y_clean)
```

```python
skf.get_support()
```

```python
x = x_clean.iloc[:, skf.get_support()]
```

```python
x
```

### Feature scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
pd.DataFrame(x_scaled).describe()
```

### cross validation

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y_clean, random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

#### Build the model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier(random_state=0)
```

```python
clf.fit(x_train, y_train)
```

### Performance analysis

```python
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
y_pred = clf.predict(x_test)
```

```python
accuracy_score(y_pred, y_test)
```

```python
new = [[120.78, 3.24, 1185, 22.29, 28.15]]
clf.predict(scaler.transform(new))[0]
```
