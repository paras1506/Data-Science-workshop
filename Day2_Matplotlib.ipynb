```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([12,23,34,45,56])
y = np.array([15,21,37,42,59])
z = np.array([87,75,54,36,25])
```

# Lineplot

```python
plt.title('Line Plot')
plt.xlabel('First value')
plt.ylabel('second value')
plt.plot(x,y, color = 'r', marker = 'v' , mfc = 'black', ls ='--', lw = 2, ms =11);

#marker =>  >,<,o, ^, D,v     (mfc = marker fint color, ls = line space, lw = line width, ms=marker size)
```

```python
plt.title('second line plot')
plt.xlabel('x label')
plt.ylabel('z label')
plt.plot(x,z , color = 'orange', marker = 'o', mfc='red', ls= '--', label = '2023');
```

```python
plt.title('second line plot')
plt.xlabel('x label')
plt.ylabel('z label')
plt.plot(x,z , color = 'orange', marker = 'o', mfc='red', ls= '--', label = '2022');
plt.plot(x,y, color = 'r', marker = 'v' , mfc = 'black', ls ='--', lw = 2, ms =11, label= '2023' );
plt.legend();
plt.grid();
```

# subplot

```python
plt.figure(figsize=(16,9))
plt.subplot(2,2,1)     #(width, height, positon)
plt.title('Line Plot')
plt.xlabel('First value')
plt.ylabel('second value')
plt.plot(x,y, color = 'r', marker = 'v' , mfc = 'black', ls ='--', lw = 2, ms =11);

plt.subplot(2,2,2)
plt.title('second line plot')
plt.xlabel('x label')
plt.ylabel('z label')
plt.plot(x,z , color = 'orange', marker = 'o', mfc='red', ls= '--', label = '2022');
plt.plot(x,y, color = 'r', marker = 'v' , mfc = 'black', ls ='--', lw = 2, ms =11, label= '2023' );
plt.legend();
plt.grid();
```

# Bar plot

```python
courses = ['java', 'C', 'Python', 'R','Kotlin']
students = [67,70,74,45,50]
```

```python
plt.title('Bar Plot')
plt.xlabel('courses')
plt.ylabel('No. of students')
plt.bar(courses, students , color = 'pink' , width = 0.6, edgecolor = 'blue', linewidth = 2 ,
         ls = '--', alpha = 0.7);

#alpha => makes color faint, default value is 1, value range is [0,1]
```

```python
plt.title('Bar Plot')
plt.xlabel('courses')
plt.ylabel('No. of students')
plt.bar(courses, students , color = ['pink', 'red','skyblue','cyan', 'yellow'] , width = 0.6, edgecolor = 'blue', linewidth = 2 ,
         ls = '--', alpha = 0.4);
```

# Scatter plot

```python
plt.title('Scatter plot')
plt.xlabel('courses')
plt.ylabel('No. of students')
plt.scatter(x, y, marker = 'o');
```

# Box Plot


```python
marks = [67,63,62,69,59,66,61,70,81]
```

```python
plt.boxplot(marks);

# box is called as interquandrantile range
```

```python
Q3 = np.percentile(marks, 75)
Q3
```

```python
Q1 = np.percentile(marks, 25)
Q1
```

```python
IQR = Q3 - Q1
IQR
```

```python
Upper_bound = Q3 + 1.5* IQR
Upper_bound
```

```python
Lower_bound = Q1 - 1.5*IQR
Lower_bound
```
