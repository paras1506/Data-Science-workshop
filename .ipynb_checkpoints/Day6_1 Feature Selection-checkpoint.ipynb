# Feature Selection

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('wisc_bc_data.csv')
```

```python
df
```

```python
df.shape
```

#### Seperate the input ans output

```python
x = df.drop(['id', 'diagnosis'], axis =1)
y = df['diagnosis']
```

```python
y.value_counts()
```

```python
x.shape
```

#### Feature Selection

```python
from sklearn.feature_selection import SelectKBest, chi2
```

```python
skf = SelectKBest(k =10, score_func= chi2)
```

```python
skf.fit(x,y)
```

```python
skf.get_support()
```

```python
x_new = x.iloc[:,skf.get_support()]
```

```python
x_new.shape
```

##### data should be in numerical format only

### Cross validation

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(x_new, y, random_state=0)
```

```python
x_train.shape, x_test.shape
```

### Build model

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=0)
clf.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = clf.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
df = pd.read_csv('mtcars.csv')
df
```

```python
x = df.drop('mpg', axis=1)
y = df['mpg']
```

```python
x.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE    # recursive feature selection method 
```

```python
reg = LinearRegression()
```

#### feature selection

```python
rfe = RFE(estimator=reg, n_features_to_select=3)
```

```python
rfe.fit(x,y)
```

```python
rfe.get_support()
```

```python
x_new = x.iloc[:,rfe.get_support()]
```

```python
x_new
```

## cross validation 

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test = train_test_split(x_new, y, random_state=0)
```

### model tarin

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_test, y_test)
```
