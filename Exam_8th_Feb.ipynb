```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1. Read the file in DataFrame

df = pd.read_csv('cricket.csv')
```

```python
df
```

```python
#2. List the name of cricketer and their respective runs
df.loc[:,['name','runs']]
```

```python
#3. Find total wickets taken by them
df['wickets'].sum()
```

```python
# 4. Find average of catches taken

df['catches'].mean()
```

```python
# 5. Find the name of wicketkeeper
df[df['stumpings'] > 0]
```

```python
# 6. print the name of bowler who played highest number of matches
bowler= df[df['wickets'] > 0]
bowler_match_high = bowler[bowler['matches'] == bowler['matches'].max()]
bowler_match_high
```

```python
#7. Find average of all the bowlers
df['bowling_average']=df['runs']/df['wickets']
df['bowling_average']
```

```python
#8. Find name of the bowler with least bowling average.

least_BowlingAvg=df.loc[df['bowling_average'].idxmin(), 'name']
least_BowlingAvg
```

```python
#9 Print information about all players whose catches per match ratio is 
# greater than 0.5 store it in a csv file

# . Print information about all players whose catches per match ratio is 
# greater than 0.5 store it in a csv file.

df['ratio']=df['catches']/df['matches']
df['catches_per_match_ratio'] = df['catches'] / df['matches']

# Filter players with catches per match ratio greater than 0.5
players = df[df['catches_per_match_ratio'] > 0.5]
players
```

```python
plt.figure(figsize=(15, 6))
plt.bar(df['name'], df['runs'], color='skyblue')
plt.xlabel('Names')
plt.ylabel('Runs')
```
